# 퍼스널컬러 측정 시스템 기술 검토서

## 1. 검토 목적

이 문서는 현재 퍼스널컬러 측정 시스템의 정확도, 구조적 한계, 해석 시 주의점, 그리고 개선 방향을 기술 검토 관점에서 정리한 문서다.  
핵심 질문은 다음과 같다.

- 현재 구조는 어느 정도 타당한가
- 어떤 부분이 실제 퍼스널컬러 진단을 단순화하고 있는가
- 어떤 구현상 특징이 결과에 큰 영향을 주는가
- 정확도를 높이려면 무엇을 바꿔야 하는가

---

## 2. 총평

이 시스템은 설명 가능성과 구현 단순성 면에서는 꽤 잘 정리된 편이다.  
얼굴 랜드마크로 ROI를 고정하고, 대표 부위 색을 추출해, 사전에 정의된 시즌 프로필과 비교한다는 점에서 구조가 투명하다.  
또한 결과 화면에서 측정값, ROI, 상위 시즌 점수를 보여주므로 사용자가 결과를 납득하기 쉽다.

하지만 정확도 관점에서는 분명한 한계가 있다.  
왜냐하면 퍼스널컬러를 결정하는 실제 요소를 매우 제한된 색 샘플 4개로 축약하고 있고, 최종 결합 단계에서 설문 비중이 항상 사진보다 더 크기 때문이다.

즉 이 시스템은 "정교한 이미지 기반 퍼스널컬러 추론기"라기보다,

`설명 가능한 규칙 기반 1차 진단기`

로 보는 것이 가장 타당하다.

---

## 3. 정확도에 유리한 요소

### 3.1 랜드마크 기반 ROI 고정

무작정 얼굴 중앙을 자르는 것이 아니라 랜드마크로 ROI를 잡기 때문에, 부위 일관성이 어느 정도 확보된다.  
이것은 단순한 이미지 크롭보다 훨씬 낫다.

### 3.2 LAB 거리 사용

팔레트 매칭에 RGB가 아니라 LAB 거리(Delta E)를 쓰는 것은 합리적이다.  
RGB 차이보다 사람의 지각에 더 가까운 비교가 가능하다.

### 3.3 네 축 모델

온도, 명도, 선명도, 대비라는 4축은 퍼스널컬러 설명 체계와 잘 맞는다.  
즉 시스템이 단지 색 하나만 보는 것이 아니라, 스타일 축을 분리해서 해석한다는 점은 장점이다.

### 3.4 품질 점수의 존재

노출, 좌우 대칭, 부위 구분도, 얼굴 크기를 별도 점수화한 것은 좋은 설계다.  
사진 상태가 나쁜 경우 분석 신뢰도를 약화시키려는 의도가 분명하기 때문이다.

---

## 4. 정확도에 불리한 핵심 한계

### 4.1 ROI가 너무 적다

현재 직접 반영되는 대표색은 사실상 4개다.

- 피부
- 머리
- 눈동자
- 입술

이것만으로 얼굴 전체의 인상을 완전히 대표한다고 보기 어렵다.  
특히 아래 요소들이 빠져 있다.

- 눈썹 색
- 피부 홍조 분포
- 턱선과 목의 톤 차이
- 눈 흰자 대비
- 얼굴 중앙과 외곽의 명도 차이

즉 현재 구조는 "대표색 기반 요약"이지 "전체 인상 분석"은 아니다.

### 4.2 평균색만 사용한다

ROI 내부를 평균 RGB로만 처리하면 분포 정보가 사라진다.

예를 들어 피부 ROI 안에서

- 홍조가 부분적으로 강한지
- 그림자가 섞였는지
- 하이라이트가 있는지
- 메이크업이 한쪽에만 있는지

를 구분하지 못한다.

이 때문에 평균색은 간단하지만 실제 얼굴의 복합적 색 구조를 크게 잃는다.

### 4.3 조명과 화이트밸런스 보정이 없다

이 시스템은 조명 보정, 화이트밸런스 보정, 색온도 보정이 사실상 없다.  
노출 적정성은 보지만 색상 캐스트 자체를 교정하지는 않는다.

예를 들어

- 노란 실내등
- 푸른 형광등
- 한쪽 측면광
- 휴대폰 카메라 자동 보정

이 들어오면 웜/쿨 추정이 쉽게 흔들릴 수 있다.

### 4.4 메이크업과 렌즈 영향이 크다

현재 측정 부위 중 입술과 눈동자는 외부 연출 영향을 많이 받는다.

- 립 틴트
- 컬러 렌즈
- 염색 머리
- 베이스 메이크업

이런 요소가 있으면 자연 고유색이 아니라 연출색을 읽게 된다.

즉 실제 퍼스널컬러보다 현재 스타일링 정보를 일부 읽을 가능성이 높다.

### 4.5 설문 비중이 지나치게 높다

이 시스템의 가장 큰 구조적 특성이다.

- 사진 가중치: 최대 0.42
- 설문 가중치: 최소 0.58

즉 사진 품질이 매우 좋아도 사진이 최종 결과의 절반을 넘지 못한다.  
반대로 설문은 항상 과반을 차지한다.

이 말은 결국 사용자의 자기 인식, 취향, 기억, 주관적 경험이 결과에 매우 큰 영향을 준다는 뜻이다.

퍼스널컬러 진단에서는 이것이 장점이 될 수도 있지만, "사진 기반 자동 진단기"라는 기대와는 다르다.

### 4.6 `decisionType`가 사실상 고정된다

코드상 `decisionType`은 `photo`, `questionnaire`, `hybrid` 중 하나인데, 계산식을 보면 사진 가중치가 0.55 이상이 될 수 없다.  
그 결과 실제로는 `photo`가 절대 나오지 않고, 대부분 `questionnaire`가 된다.

이는 UI 메시지와 시스템 실제 동작 사이에 차이를 만들 수 있다.

즉 사용자는 하이브리드처럼 느끼지만 내부적으로는 설문 주도 구조다.

---

## 5. 해석상 주의해야 할 점

### 5.1 결과는 절대적 진단이 아니다

이 프로그램 결과는 "당신의 절대 퍼스널컬러"라기보다,

- 현재 촬영 환경
- 현재 스타일링
- 현재 응답

을 바탕으로 계산된 상대적 시즌 추천이다.

### 5.2 1위와 2위가 가까우면 확정적으로 보면 안 된다

이 시스템은 1위와 2위 점수 차이를 confidence에 반영한다.  
따라서 상위 두 시즌이 비슷하게 나오면, 단일 시즌 확정보다 인접 시즌군으로 이해하는 것이 더 적절하다.

### 5.3 사용자 설문이 잘못되면 결과도 흔들린다

특히 아래 문항들은 자기 인식 오류의 영향을 받기 쉽다.

- 선명한 색이 어울리는지
- 뮤트 색이 어울리는지
- 고대비가 어울리는지
- 깊은 색이 안정적인지

이런 질문은 객관 관찰보다 주관 판단 비중이 높기 때문이다.

## 6. 개선 우선순위 제안

아래는 실제 정확도 개선 효과가 큰 순서대로 정리한 제안이다.

### 6.1 1순위: 조명 보정 추가

가장 먼저 필요한 것은 색상 교정이다.

추천 방향:

- 얼굴 주변 회색 기준점 또는 화이트 레퍼런스 유도
- 자동 화이트밸런스 보정
- 색온도 보정
- 그림자/하이라이트 제거를 위한 간단한 정규화

이 개선은 웜/쿨 오판을 크게 줄일 수 있다.

### 6.2 2순위: ROI 확장

현재 4대표색 구조를 확장할 필요가 있다.

추가 후보:

- 눈썹
- 목 피부
- 턱선
- 이마 중앙
- 눈 흰자 주변
- 코 옆 붉은기 영역

이렇게 하면 평균색 4개보다 훨씬 안정적인 얼굴 프로필을 얻을 수 있다.

### 6.3 3순위: 평균색 대신 분포 특징 추가

평균 RGB만 쓰지 말고 다음도 함께 쓰는 것이 좋다.

- 채도 분산
- 명도 분산
- ROI 내 분위수 색상
- 그림자 제거 후 평균색
- 좌우 차이 편차

이렇게 하면 한 부위 내 복합 정보를 잃지 않을 수 있다.

### 6.4 4순위: 설문 비중 재조정

현재는 설문이 항상 우세하다.  
만약 "사진 기반 진단" 성격을 강화하려면 가중치를 바꿔야 한다.

예시 방향:

- 사진 품질이 높을수록 사진 비중을 0.6~0.7까지 올리기
- 설문은 보정 신호로만 유지
- 설문 문항별 신뢰도도 따로 받기

이렇게 해야 truly hybrid 구조가 된다.

### 6.5 5순위: 메이크업/렌즈/염색 체크 문항 추가

사진 분석에 영향을 줄 수 있는 조건을 먼저 묻는 것도 중요하다.

예시:

- 립 제품 사용 여부
- 컬러 렌즈 여부
- 염색 여부
- 베이스 메이크업 여부

이 정보를 받아 특정 ROI의 가중치를 낮출 수 있다.

### 6.6 6순위: 시즌 경계 처리 강화

현재는 최종적으로 1위/2위를 제시하지만, 경계형 처리 로직을 더 강화할 수 있다.

예시:

- top1과 top2 차이가 작으면 "혼합형"으로 표시
- family는 같고 세부 시즌만 다르면 family 우선 제안
- 봄-가을, 여름-겨울 등 이웃 축 관계를 설명에 포함

이렇게 하면 실사용 해석 품질이 좋아진다.

`실시간 얼굴 랜드마크로 색 샘플링 위치를 정하고, 대표 부위 색을 12계절 팔레트 및 설문 축과 비교해 시즌을 추천하는 설명 가능한 규칙 기반 퍼스널컬러 진단 시스템`

여기서 핵심 포인트는 세 가지다.

1. 설명 가능하다.
2. 규칙 기반이다.
3. 설문이 강하게 반영된다.

## 9. 최종 판단

기술적으로 이 프로그램은 "잘 정리된 1차 진단기"로서는 충분히 의미가 있다.  
특히 시연용, 교육용, 인터랙티브 분석 도구로는 강점이 있다.

하지만 전문 진단 수준의 정밀도를 기대하려면 아직 부족한 부분이 많다.

그 이유는 아래 세 줄로 정리할 수 있다.

- 얼굴 색 정보를 너무 적은 ROI로 압축하고 있다.
- 촬영 환경 보정이 약하다.
- 최종적으로는 설문이 사진보다 더 강하다.

따라서 현재 버전은 "자동 퍼스널컬러 확정 시스템"보다는

`사진 + 설문 기반 시즌 추천 보조 도구`

로 보는 것이 가장 현실적이다.


# 퍼스널컬러 측정 시스템 상세 보고서

## 1. 문서 목적

이 문서는 현재 프로젝트에 포함된 퍼스널컬러 측정 프로그램이 어떤 방식으로 동작하는지, 처음 보는 사람도 전체 구조를 이해할 수 있도록 설명하기 위한 기술 보고서다.  
설명 범위는 다음을 포함한다.

- 전체 실행 흐름
- 12계절 퍼스널컬러 체계가 코드에 반영된 방식
- 얼굴 인식과 색상 추출 방식
- 볼, 눈동자, 입술, 헤어라인 등의 수치화 방식
- 설문 문항이 얻으려는 정보와 가중치
- 사진 결과와 설문 결과가 최종 판정으로 합쳐지는 구조
- 시스템의 실제 성격과 해석 시 주의점

이 시스템은 흔히 말하는 "생성형 AI가 얼굴 전체를 보고 추론하는 모델"이라기보다, 얼굴 특정 부위의 색을 수치화하고 이를 사전에 정의된 12계절 팔레트와 설문 응답에 대입하여 판정하는 규칙 기반 하이브리드 시스템에 가깝다.

## 2. 시스템 한눈에 보기

이 프로그램은 React + TypeScript 기반의 브라우저 앱이다. 사용자는 웹캠으로 자신의 얼굴을 촬영하고, 이어서 8개 설문 문항에 응답한다. 시스템은 두 종류의 신호를 만든다.

1. 사진 기반 신호
2. 설문 기반 신호

그 뒤 두 신호를 합쳐 최종 퍼스널컬러 시즌을 판정한다.

핵심 구조는 다음과 같다.

1. 카메라에서 얼굴을 인식한다.
2. 얼굴 랜드마크를 바탕으로 샘플링 부위를 정한다.
3. 피부, 머리, 눈동자, 입술의 평균 색을 추출한다.
4. 추출한 색을 온도, 명도, 선명도, 대비 같은 축으로 변환한다.
5. 12계절 각각의 프로필과 비교해 사진 점수를 만든다.
6. 설문 응답을 같은 축으로 정규화해 설문 점수를 만든다.
7. 두 점수를 합쳐 최종 시즌과 신뢰도를 계산한다.

## 3. 실행 흐름 상세

### 3.1 단계 전환 구조

앱은 다음 5단계로 동작한다.

- `intro`: 시작 화면
- `photo`: 사진 분석 단계
- `questionnaire`: 설문 단계
- `fusing`: 사진과 설문을 합치는 단계
- `result`: 최종 결과 화면

즉, 이 시스템은 "사진만 찍고 바로 결과"가 아니라 반드시 설문 단계를 거치도록 설계되어 있다.

### 3.2 실제 사용자 흐름

사용자가 앱에 들어오면 먼저 시스템 설명을 본 뒤 분석을 시작한다.  
이후 카메라가 켜지고, 얼굴이 프레임 안에 잡히면 시스템이 랜드마크를 실시간으로 추적한다.  
사용자가 촬영 버튼을 누르면 정지된 프레임에서 색을 추출하고, 그 결과를 임시 저장한다.  
그 다음 설문으로 넘어가며, 모든 설문이 끝나면 사진 결과와 설문 결과를 합쳐 최종 결과가 출력된다.

## 4. 12계절 퍼스널컬러 체계가 반영된 방식

### 4.1 코드에 정의된 12개 시즌

이 프로그램은 4개 대분류와 12개 세부 시즌을 사용한다.

#### 봄(Spring)

- 라이트 스프링
- 트루 스프링
- 브라이트 스프링

#### 여름(Summer)

- 라이트 서머
- 트루 서머
- 소프트 서머

#### 가을(Autumn)

- 소프트 어텀
- 트루 어텀
- 다크 어텀

#### 겨울(Winter)

- 다크 윈터
- 트루 윈터
- 브라이트 윈터

### 4.2 각 시즌에 들어 있는 정보

각 시즌은 코드 내부에 하나의 프로필 객체로 정의되어 있으며, 시즌마다 다음 정보가 들어 있다.

- 시즌 ID
- 한국어 이름
- 영어 이름
- 대분류 family
- 요약 톤 메모
- 4축 특성값
- 엑셀 기반 워크북 통계
- 24색 팔레트

### 4.3 4축 특성값의 의미

각 시즌은 아래 네 개의 축으로 표현된다.

- `temperature`: 웜/쿨
- `lightness`: 라이트/딥
- `clarity`: 클리어/뮤트
- `contrast`: 고대비/저대비

예를 들어 브라이트 윈터는 대체로 다음처럼 표현된다.

- 쿨 성향
- 중간 이상 명도
- 매우 높은 선명도
- 매우 높은 대비

반대로 소프트 서머는 다음처럼 표현된다.

- 쿨 성향
- 중간 명도
- 매우 뮤트
- 낮거나 중간 수준 대비

즉, 프로그램은 사람의 얼굴을 "봄 같다"처럼 직접 판단하는 대신, 먼저 얼굴을 네 축의 수치로 바꾼 다음 각 시즌 프로필과 얼마나 가까운지를 본다.

### 4.4 엑셀 팔레트의 역할

프로젝트에는 `personal_color_12season_24palette.xlsx`라는 파일이 있고, 코드에는 이 데이터를 바탕으로 만든 12개 시즌의 24색 팔레트가 하드코딩되어 있다.

중요한 점은, 실행 시점에 엑셀 파일을 읽는 것이 아니라 이미 정리된 값이 코드에 들어 있다는 점이다.  
즉 "엑셀 기반"이지만 런타임 계산은 코드 내부 팔레트를 사용한다.

## 5. 얼굴 인식 방식

### 5.1 사용 라이브러리

얼굴 인식에는 `@mediapipe/tasks-vision`의 `FaceLandmarker`가 사용된다.

이 말은 곧 다음을 뜻한다.

- 자체적으로 얼굴 인식 모델을 학습한 시스템이 아니다.
- Google MediaPipe가 제공하는 사전학습 얼굴 랜드마크 모델을 사용한다.
- 브라우저 안에서 실행된다.

### 5.2 모델 로딩 방식

모델은 외부 경로에서 불러온다.

- WASM 실행 리소스: jsDelivr CDN
- 모델 파일: Google Cloud Storage

즉 초기 로딩 시 네트워크 연결이 필요하다.

### 5.3 동작 설정

설정값은 아래와 같다.

- `runningMode`: `VIDEO`
- `numFaces`: `1`
- `minFaceDetectionConfidence`: `0.35`
- `minFacePresenceConfidence`: `0.35`
- `minTrackingConfidence`: `0.35`

실행은 GPU를 우선 시도하고, 실패하면 CPU로 대체한다.

### 5.4 이 단계에서 실제로 얻는 것

얼굴 인식 단계에서 프로그램은 사람의 감정이나 나이, 성별을 추론하지 않는다.  
이 단계의 목적은 오직 하나다.

`얼굴의 주요 위치 좌표를 얻어 색을 샘플링할 정확한 영역을 정하는 것`

즉 랜드마크는 얼굴 이해용이 아니라 색 측정 위치 지정용이다.

## 6. 얼굴에서 어떤 부위를 측정하나

### 6.1 실제 측정 대상

이 프로그램이 최종적으로 직접 사용하는 대표 색상은 4개다.

- 피부
- 머리
- 눈동자
- 입술

보다 정확히 말하면 샘플링은 6개 ROI에서 일어나지만, 결과는 4개 대표색으로 합쳐진다.

#### 샘플링 ROI 6개

- 왼쪽 볼
- 오른쪽 볼
- 왼쪽 눈동자
- 오른쪽 눈동자
- 입술 중심
- 헤어라인 상단

#### 최종 대표색 4개

- 피부 = 왼쪽 볼 + 오른쪽 볼 평균
- 눈동자 = 왼쪽 눈동자 + 오른쪽 눈동자 평균
- 머리 = 헤어라인 영역 색
- 입술 = 입술 중심 색

### 6.2 랜드마크 기준점

ROI 중심을 잡는 데 사용되는 랜드마크는 다음과 같다.

- 왼쪽 볼: `205`
- 오른쪽 볼: `425`
- 이마 기준점: `10`
- 입술 중심 평균: `13, 14, 78, 308`
- 왼쪽 홍채: `468, 469, 470, 471, 472`
- 오른쪽 홍채: `473, 474, 475, 476, 477`

### 6.3 ROI 크기

ROI는 얼굴 박스 크기에 비례해서 정해진다.

- 볼: 얼굴 너비 14%, 얼굴 높이 10%
- 눈동자: 얼굴 너비 6%, 얼굴 높이 5%
- 입술: 얼굴 너비 14%, 얼굴 높이 6%
- 헤어라인: 얼굴 너비 18%, 얼굴 높이 8%

이 방식의 장점은 얼굴이 크든 작든 비슷한 비율로 샘플링한다는 점이다.

### 6.4 눈썹은 왜 없나

질문에서 눈썹을 예로 들었지만, 현재 구현은 눈썹을 독립 ROI로 측정하지 않는다.  
머리카락 색은 헤어라인 상단으로 대체하고, 눈 관련 정보는 홍채 중심 색으로 대표한다.

즉 시스템은 "눈썹색"이나 "볼 홍조 분포" 같은 세밀한 미적 특징을 따로 정량화하지 않는다.

## 7. 색 추출 방식

### 7.1 픽셀 평균 방식

각 ROI에 대해 프로그램은 해당 사각형의 픽셀을 읽고, 유효 픽셀의 RGB 평균을 계산한다.

처리 방식은 단순 평균이다.

- 알파값이 180 미만인 픽셀은 제외
- 나머지 픽셀의 R, G, B를 각각 합산
- 픽셀 수로 나눠 평균 계산

즉, 텍스처 분석이나 분산 분석 없이 "대표 평균색"만 본다.

### 7.2 왜 평균색을 쓰는가

평균색은 다음 장점이 있다.

- 계산이 빠르다
- 설명 가능성이 높다
- 팔레트 비교에 적합하다

하지만 단점도 있다.

- 조명 얼룩에 민감하다
- 메이크업 영향을 크게 받을 수 있다
- 그림자와 하이라이트가 섞이면 대표성이 떨어질 수 있다

---

## 8. 색상 수치화: RGB, LAB, HSL

### 8.1 RGB

RGB는 가장 기본적인 색 표현이다.

- R: 빨강
- G: 초록
- B: 파랑

이 값은 주로 화면 표시와 원시 색 보관용으로 사용된다.

### 8.2 LAB

LAB는 사람의 색 지각과 비교적 잘 맞는 색 공간이다.

- `L`: 밝기
- `a`: 초록(-) ↔ 빨강(+)
- `b`: 파랑(-) ↔ 노랑(+)

이 프로그램은 LAB를 시즌 팔레트와의 거리 계산에 사용한다.  
즉 "이 얼굴 색이 이 시즌 팔레트와 얼마나 가까운가"를 단순 RGB 차이가 아니라 LAB 거리로 본다.

### 8.3 HSL

HSL은 해석이 쉬운 색 공간이다.

- `H`: 색상각
- `S`: 채도
- `L`: 밝기

이 프로그램에서는 특히 `S`가 중요하다.  
왜냐하면 선명함과 뮤트 성향을 채도 기반으로 계산하기 때문이다.

---

## 9. 사진에서 추출하는 핵심 특징값

사진 분석의 핵심 출력은 다음 5개다.

- `temperature`
- `lightness`
- `clarity`
- `contrast`
- `mutedScore`

### 9.1 temperature

웜/쿨 성향을 나타낸다.  
계산에는 `colorTemperatureIndex`라는 간이 지표가 사용된다.

대략적 아이디어는 이렇다.

- 빨강이 강하고 파랑이 약하면 웜 쪽
- 파랑이 강하면 쿨 쪽

최종 온도는 가중 평균으로 계산된다.

- 피부: 0.45
- 입술: 0.25
- 머리: 0.15
- 눈동자: 0.15

즉 피부색의 영향이 가장 크다.

### 9.2 lightness

얼굴 전반의 밝고 어두운 성향을 뜻한다.

가중치는 다음과 같다.

- 피부: 0.45
- 입술: 0.20
- 눈동자: 0.15
- 머리: 0.20

결과는 `-1 ~ 1` 범위로 정규화된다.

- 양수: 라이트 성향
- 음수: 딥 성향

### 9.3 clarity

선명한 색이 받는지, 탁한 색이 받는지를 나타낸다.

이 값은 평균 채도 기반으로 계산된다.

- 피부 채도: 0.40
- 입술 채도: 0.25
- 눈동자 채도: 0.20
- 머리 채도: 0.15

이 평균 채도를 `-1 ~ 1`로 변환한 값이 clarity다.

- 양수: 클리어
- 음수: 뮤트

### 9.4 mutedScore

뮤트 정도를 별도로 나타내는 보조값이다.

- `mutedScore = 1 - averageSaturation`

즉 채도가 낮을수록 mutedScore는 커진다.

### 9.5 contrast

얼굴 내 부위 간 밝기 차이를 의미한다.

- 피부, 머리, 눈동자, 입술의 밝기 중
- 가장 밝은 값과 가장 어두운 값의 차이를 계산

결과적으로

- 값이 크면 고대비 얼굴
- 값이 작으면 저대비 얼굴

로 해석된다.

## 10. 사진 품질 평가

이 시스템은 사진 자체의 품질도 점수화한다.  
이 점수는 사진 기반 신호의 신뢰도를 나타내며 최종 confidence에도 반영된다.

### 10.1 exposure

피부 평균 밝기가 너무 어둡거나 너무 밝지 않은지 본다.

- 이상적인 밝기 중심값: `0.62`

즉 노출이 적절할수록 점수가 높다.

### 10.2 symmetry

왼쪽 볼과 오른쪽 볼의 LAB 거리 차이를 본다.

좌우 볼 색이 지나치게 다르면

- 조명이 한쪽만 강하거나
- 그림자가 지거나
- 측정 안정성이 낮다고 판단

한다.

### 10.3 distinctness

피부와 머리/눈/입술이 충분히 구분되는지 본다.

이 값이 너무 낮으면 얼굴 부위들이 전부 비슷하게 측정되었다는 뜻이므로 특징 추출력이 낮다고 본다.

### 10.4 faceSize

얼굴이 프레임 안에서 충분히 크게 잡혔는지를 의미한다.

너무 멀리서 찍으면 세부 색 측정 신뢰도가 떨어지기 때문이다.

### 10.5 overall 사진 품질

최종 사진 품질은 다음 성분을 합쳐 계산된다.

- exposure: 0.28
- symmetry: 0.22
- distinctness: 0.20
- faceSize: 0.12
- 기본 상수: 0.18

즉 조명, 좌우 균형, 색 차이, 얼굴 크기를 모두 고려한 종합 품질 점수다.

## 11. 사진 기반 시즌 점수 계산 방식

각 시즌별로 사진 점수는 두 층으로 계산된다.

### 11.1 팔레트 거리 점수

피부, 머리, 눈동자, 입술의 LAB 색을 해당 시즌 24색 팔레트 전체와 비교한다.  
각 부위는 "가장 가까운 팔레트 색" 하나와의 거리만 사용한다.

부위별 가중치는 다음과 같다.

- 피부: 0.45
- 머리: 0.20
- 눈동자: 0.15
- 입술: 0.20

즉 사진 판정에서 가장 중요한 단일 색은 피부다.

### 11.2 축 특성 점수

추출된 4축 특징값이 시즌의 이상적 traits와 얼마나 가까운지 계산한다.

가중치는 다음과 같다.

- temperature: 0.38
- lightness: 0.18
- clarity: 0.24
- contrast: 0.20

### 11.3 두 점수의 합성

최종 사진 raw score는 다음처럼 계산된다.

- 팔레트 거리 점수: 68%
- 축 특성 점수: 32%

즉 이 시스템은 사진 판정에서 "직접적인 색 팔레트 유사도"를 더 강하게 본다.

## 12. 설문 구조와 문항별 의미

설문은 총 8문항이며, 모든 문항은 결국 다음 네 축으로 환산된다.

- `temperature`
- `lightness`
- `clarity`
- `contrast`

### 12.1 손목 혈관색

얻고 싶은 정보:

- 웜/쿨 베이스 추정

해석:

- 초록/올리브에 가까우면 웜 쪽
- 블루/퍼플에 가까우면 쿨 쪽

### 12.2 액세서리 반응

얻고 싶은 정보:

- 웜/쿨 베이스
- 금속 반사와 얼굴 선명감의 관계

해석:

- 골드가 살아 보이면 웜 가능성
- 실버/플래티넘이 살아 보이면 쿨 가능성

### 12.3 흰옷 반응

얻고 싶은 정보:

- 웜/쿨
- 어떤 밝기의 흰색이 받는지
- 선명한 백색이 부담스러운지

### 12.4 햇빛 반응

얻고 싶은 정보:

- 피부 베이스
- 피부 반응 패턴을 통한 대비/온도 보조 판단

### 12.5 선명한 고채도 색 반응

얻고 싶은 정보:

- 클리어 vs 뮤트
- 고대비를 소화하는지

### 12.6 뮤트 컬러 반응

얻고 싶은 정보:

- 뮤트 컬러 적합성

이 문항은 선명도 축을 반대 방향에서 확인하는 역할을 한다.

### 12.7 대비감 선호

얻고 싶은 정보:

- 얼굴이 강한 대비를 감당하는지
- 선명도와의 연계

### 12.8 깊이감 선호

얻고 싶은 정보:

- 밝은색이 안정적인지
- 중간톤이 맞는지
- 깊고 짙은 컬러가 맞는지

즉 lightness 축을 보정한다.

## 13. 설문 가중치 구조

각 보기에는 특정 축의 가중치가 들어 있다.  
이 가중치는 사람이 선택한 답을 수치 벡터로 바꾸는 역할을 한다.

핵심 해석은 다음과 같다.

- 온도는 혈관, 액세서리, 흰옷, 햇빛 반응에서 주로 결정된다.
- 선명도는 선명한 색, 뮤트 색, 대비감 문항에서 강하게 결정된다.
- 명도는 흰옷과 깊이감 문항에서 주로 결정된다.
- 대비는 선명한 색, 햇빛, 대비감 문항에서 보정된다.

즉 설문은 단순 취향 조사가 아니라, 색 반응 패턴을 네 축으로 분해하는 장치다.

## 14. 설문 점수 정규화

설문 응답을 모두 더하면 축마다 범위가 다르기 때문에, 그대로 비교하면 불공정하다.  
그래서 이 프로그램은 각 축의 총점을 그 축이 가질 수 있는 최대 절대값 합으로 나누어 `-1 ~ 1` 범위로 정규화한다.

정규화 결과 해석은 다음과 같다.

- temperature > 0: 웜
- temperature < 0: 쿨
- lightness > 0: 라이트
- lightness < 0: 딥
- clarity > 0: 클리어
- clarity < 0: 뮤트
- contrast > 0: 고대비
- contrast < 0: 저대비

즉 설문은 단순 카테고리 선택이 아니라, 최종적으로는 연속적인 벡터를 만든다.

각 시즌 프로필의 trait와 설문 축 값을 비교해 유사도를 만든다.

가중치는 다음과 같다.

- temperature: 0.38
- lightness: 0.20
- clarity: 0.25
- contrast: 0.17

이 말은 곧 설문에서는 다음이 중요하다는 뜻이다.

1. 웜/쿨
2. 선명함/뮤트
3. 명도
4. 대비

특히 온도축의 영향력이 가장 크다.

## 16. 사진과 설문의 결합 방식

이 시스템의 가장 중요한 구현적 특징은 최종 결과가 사진보다 설문에 더 크게 좌우된다는 점이다.

### 16.1 사진 가중치

사진 가중치는 사진 품질에 따라 바뀐다.

- 최소: 0.18
- 최대: 0.42

### 16.2 설문 가중치

설문 가중치는 항상 다음과 같다.

- 최소: 0.58
- 최대: 0.82

즉 구조상 설문이 항상 더 강하다.

### 16.3 의미

따라서 이 시스템은 "사진 판정기에 설문이 조금 붙은 것"이 아니라,

`설문 중심 판정기에 사진이 보조 증거로 결합되는 구조`

라고 보는 편이 더 정확하다.

## 17. 최종 confidence와 일치도

최종 confidence는 단순히 1위 점수만으로 정해지지 않는다.

반영 요소는 다음과 같다.

- 1위 시즌 점수
- 1위와 2위 점수 차이
- 사진 품질
- 사진과 설문 신호의 일치도

### 17.1 일치도 판단

일치도는 다음 두 조건을 본다.

- 사진과 설문의 웜/쿨 방향이 같은가
- 사진 1위 시즌 family와 설문 1위 season family가 같은가

판정:

- 둘 다 같으면 `high`
- 하나만 같으면 `medium`
- 둘 다 다르면 `low`

즉 시스템은 사진과 설문이 서로 같은 방향을 가리키는지까지 반영한다.

## 18. 결과 화면이 보여주는 것

최종 결과 화면에서는 단순히 시즌 이름만 보여주지 않는다.  
다음 정보도 함께 제공한다.

- 최종 1위 시즌
- 2위 시즌
- confidence
- 사진 신호 요약
- 설문 신호 요약
- 두 신호의 일치도
- 추천 팔레트 일부
- 추출된 샘플 색상
- ROI별 RGB / LAB / HSL 값
- 얼굴 바운딩 박스
- 사진 품질 세부 점수
- 상위 시즌 점수 목록

즉 사용자는 "왜 이런 결과가 나왔는지"를 어느 정도 역추적할 수 있다.

## 19. 이 프로그램의 실제 성격

이 시스템을 정확히 표현하면 다음과 같다.

`얼굴 랜드마크 기반 색채 샘플링과 설문 정규화를 결합한 규칙 기반 퍼스널컬러 시즌 판정기`

중요한 특징은 아래와 같다.

- 생성형 AI 기반 전체 얼굴 해석기가 아니다.
- 메이크업 진단기나 피부 분석기는 아니다.
- 얼굴 특정 부위의 대표색에 매우 민감하다.
- 설문 비중이 높다.
- 결과 해석 가능성이 높다.

---

## 20. 이 시스템이 잘하는 것과 못하는 것

### 20.1 잘하는 것

- 구조가 명확하고 설명 가능하다.
- 각 결과를 수치와 팔레트로 뒷받침할 수 있다.
- 브라우저에서 바로 동작한다.
- 실시간 랜드마크 기반 ROI 시각화가 가능하다.

### 20.2 못하는 것

- 조명 변화에 취약하다.
- 메이크업, 렌즈, 염색 영향을 직접 교정하지 않는다.
- 눈썹, 홍조 분포, 그림자 패턴, 피부 텍스처는 반영하지 않는다.
- 사진보다 설문이 훨씬 강하기 때문에 실제 색채 신호보다 자기 인식 응답이 결과를 크게 끌 수 있다.

---

## 21. 최종 결론

이 퍼스널컬러 측정 시스템은 얼굴 전체를 의미적으로 이해하는 AI가 아니라, 얼굴의 몇 개 대표 부위에서 색을 추출하여 수치화하고, 이를 12계절 팔레트와 설문 축에 비교하는 계산형 시스템이다.

판정의 핵심은 다음 세 문장으로 요약된다.

1. 얼굴 랜드마크로 샘플링 위치를 정확히 잡는다.
2. 피부, 눈동자, 입술, 머리 색을 웜/쿨, 명도, 선명도, 대비 값으로 바꾼다.
3. 12계절 팔레트와 설문 축을 함께 비교해 최종 시즌을 정한다.

즉 이 프로그램은 "설명 가능한 퍼스널컬러 규칙 엔진"으로 이해하는 것이 가장 정확하다.
